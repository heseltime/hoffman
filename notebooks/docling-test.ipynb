````python
import os
from docling.document_converter import DocumentConverter

# Define the local PDF path (relative to the current notebook directory)
local_pdf_path = os.path.join("..", "samples", "Down The Rabbit Hole.pdf")

# Set up the DocumentConverter
converter = DocumentConverter()

# Perform the conversion
try:
    # Convert the document
    result = converter.convert(local_pdf_path)

    # Export to Markdown
    markdown_output = result.document.export_to_markdown()

    # Print the output to verify
    print(markdown_output)

    # Save the Markdown output to a file
    output_md_path = os.path.join("Down The Rabbit Hole.md")
    with open(output_md_path, "w", encoding="utf-8") as md_file:
        md_file.write(markdown_output)

    print(f"\nMarkdown content saved to: {output_md_path}")

except Exception as e:
    print(f"Error during conversion: {e}")
````
